```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters for the Mobius strip
radius = 1.0
width = 0.2
num_twists = 1.5
num_points = 200

# Generate the Mobius strip coordinates
u = np.linspace(0, 2 * np.pi, num_points)
v = np.linspace(-width / 2, width / 2, num_points)
u, v = np.meshgrid(u, v)
x = (radius + v * np.cos(u / 2) * np.cos(num_twists * u)) * np.cos(u)
y = (radius + v * np.cos(u / 2) * np.cos(num_twists * u)) * np.sin(u)
z = v * np.sin(u / 2) * np.cos(num_twists * u)

# Plotting the Mobius strip
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set plot parameters
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Mobius Strip')

# Display the plot
plt.show()
```

```python
import turtle

def draw_fractal_tree(t, branch_length, angle):
    if branch_length < 5:  # Base case: stop recursion when branch length is small
        return
    else:
        t.forward(branch_length)  # Draw the trunk
        t.left(angle)  # Turn left to draw the left subtree
        draw_fractal_tree(t, branch_length * 0.7, angle)  # Recursively draw the left subtree
        t.right(angle * 2)  # Turn right to draw the right subtree
        draw_fractal_tree(t, branch_length * 0.7, angle)  # Recursively draw the right subtree
        t.left(angle)  # Turn left to restore the original orientation
        t.backward(branch_length)  # Move backward to the starting position

# Create the turtle object and set initial parameters
my_turtle = turtle.Turtle()
my_screen = turtle.Screen()
my_screen.bgcolor("white")
my_turtle.left(90)  # Start facing upwards
my_turtle.up()  # Lift the pen
my_turtle.goto(0, -200)  # Position the turtle at the bottom

# Draw the fractal tree
my_turtle.down()  # Lower the pen
my_turtle.color("green")  # Set the color
draw_fractal_tree(my_turtle, 100, 30)  # Call the function to draw the tree

# Exit on click
my_screen.exitonclick()
```

```python
import turtle

def draw_sierpinski_triangle(t, depth, size):
    if depth == 0:
        for _ in range(3):
            t.forward(size)
            t.left(120)
    else:
        draw_sierpinski_triangle(t, depth - 1, size / 2)  # Draw the top triangle
        t.forward(size / 2)
        draw_sierpinski_triangle(t, depth - 1, size / 2)  # Draw the left triangle
        t.backward(size / 2)
        t.left(60)
        t.forward(size / 2)
        t.right(60)
        draw_sierpinski_triangle(t, depth - 1, size / 2)  # Draw the right triangle
        t.left(60)
        t.backward(size / 2)
        t.right(60)

# Create the turtle object and set initial parameters
my_turtle = turtle.Turtle()
my_screen = turtle.Screen()
my_screen.bgcolor("white")
my_turtle.up()  # Lift the pen
my_turtle.goto(-200, -200)  # Position the turtle at the bottom-left corner

# Draw the Sierpinski Triangle
my_turtle.down()  # Lower the pen
my_turtle.color("green")  # Set the color
draw_sierpinski_triangle(my_turtle, 4, 400)  # Call the function to draw the triangle

# Exit on click
my_screen.exitonclick()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot_set(width, height, x_min, x_max, y_min, y_max, max_iterations):
    x = np.linspace(x_min, x_max, width)
    y = np.linspace(y_min, y_max, height)
    X, Y = np.meshgrid(x, y)
    c = X + 1j * Y
    z = np.zeros_like(c, dtype=complex)
    fractal = np.zeros(c.shape, dtype=int)

    for i in range(max_iterations):
        mask = np.logical_and(fractal == 0, np.abs(z) < 2)
        z[mask] = z[mask] * z[mask] + c[mask]
        fractal += mask

    return fractal

# Set the parameters for the Mandelbrot Set
width = 800
height = 600
x_min = -2.5
x_max = 1
y_min = -1.5
y_max = 1.5
max_iterations = 100

# Generate the Mandelbrot Set
fractal = mandelbrot_set(width, height, x_min, x_max, y_min, y_max, max_iterations)

# Plot the Mandelbrot Set
plt.figure(figsize=(10, 8))
plt.imshow(fractal.T, cmap='hot', extent=(x_min, x_max, y_min, y_max))
plt.title('Mandelbrot Set')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.colorbar(label='Iterations')
plt.show()
```

```python
import turtle

def draw_fibonacci_spiral(length, n):
    a, b = 0, 1
    for _ in range(n):
        turtle.forward(length * b)
        turtle.left(90)
        a, b = b, a + b

# Setup turtle
turtle.speed(0)
turtle.penup()
turtle.goto(-200, 0)
turtle.pendown()

# Draw Fibonacci spiral
draw_fibonacci_spiral(2, 10)

# Hide turtle
turtle.hideturtle()

# Exit on click
turtle.exitonclick()
```

```python
import turtle

def draw_kite():
    for _ in range(2):
        turtle.forward(100)
        turtle.right(72)
        turtle.forward(100)
        turtle.right(108)

def draw_dart():
    for _ in range(2):
        turtle.forward(100)
        turtle.right(36)
        turtle.forward(100)
        turtle.right(144)

def draw_penrose_tiling(level, size):
    if level == 0:
        return

    turtle.fillcolor("black")
    turtle.begin_fill()
    draw_kite()
    turtle.end_fill()

    turtle.fillcolor("gray")
    turtle.begin_fill()
    for _ in range(5):
        draw_dart()
    turtle.end_fill()

    turtle.penup()
    turtle.right(36)
    turtle.forward(size)
    turtle.left(36)
    turtle.pendown()

    draw_penrose_tiling(level - 1, size * 0.382)

# Setup turtle
turtle.speed(0)
turtle.penup()
turtle.goto(-200, -200)
turtle.pendown()

# Draw Penrose tiling
draw_penrose_tiling(5, 300)

# Hide turtle
turtle.hideturtle()

# Exit on click
turtle.exitonclick()
```

```python
import math

def fibonacci(n):
    sequence = [1, 1]
    for _ in range(2, n + 1):
        sequence.append(sequence[-1] + sequence[-2])
    return sequence

def check_golden_ratio(n):
    fib_sequence = fibonacci(n)
    ratio = fib_sequence[-1] / fib_sequence[-2]
    golden_ratio = (1 + math.sqrt(5)) / 2
    return abs(ratio - golden_ratio) < 0.0001

n = 1
while not check_golden_ratio(n):
    n += 1

print(f"The golden ratio is approximately {fibonacci(n)[-1] / fibonacci(n - 1)[-1]}")
```
